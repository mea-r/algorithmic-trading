**Download daily price data**

```python
def get_data():
    tickers = ["SPY", "TSLA", "QUBT", "IBM", "PLUG", "NOW", "RHM.DE", "NVDA", "AAPL", "AMZN", "MSFT", "GOOGL", "META", "NFLX", "AMD", "INTC", "ORCL", "SAP.DE", "ASML.AS", "TSM", "BABA", "JD", "PDD", "SHOP.TO", "PAA", "PYPL", "V", "MA", "JPM", "BAC", "GS", "BRK-B", "XOM", "CVX", "COP", "KO", "PEP", "MCD", "NKE", "DIS", "WMT", "COST", "UNH", "SIE.DE", "VOW3.DE", "MC.PA", "QQQ", "IWM", "TLT", "GLD"]
    return yf.download(tickers, start = "2015-01-01", end = "2025-01-01", auto_adjust = False)
#data.to_csv("raw.csv")
```

**Clean data**

```python
def remove_NA(df):
    for col in df.columns:
        df = df[df[col].notna()]
    return df

def convert_columns_to_numeric(df):
    return df.apply(pd.to_numeric)

def convert_timestamp_to_datetime(df):
    return df

def remove_duplicates(df):
    return df.drop_duplicates()
```

**Convert to parquet file format**

```python
def convert_to_parquet(df, file_name):
    df.to_parquet(file_name)
```

**Program startup**

```python
def main():
    df_tickers_OHLCV_raw = get_data()
    df_tickers_OHLCV_clean = remove_duplicates(convert_timestamp_to_datetime(convert_columns_to_numeric(remove_NA(df_tickers_OHLCV_raw))))
    #df_tickers_OHLCV_clean.to_csv("50_assets_test_data.csv")
    print(df_tickers_OHLCV_clean)
    convert_to_parquet(df_tickers_OHLCV_clean, "clean_prices_50_assets.parquet")
main()
```
