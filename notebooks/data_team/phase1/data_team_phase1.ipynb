**Import relevant libraries**

```python
import yfinance as yf
import pandas as pd
```

**Download daily price data**

```python
def get_data():
    tickers = ["SPY", "TSLA", "QUBT", "IBM", "PLUG", "NOW", "RHM.DE", "NVDA", "AAPL", "AMZN"]
    return yf.download(tickers, start = "2015-01-01", end = "2025-01-01", auto_adjust = True)
#data.to_csv("raw.csv")
```

**Clean data**

```python
def remove_NA(df):
    for col in df.columns:
        df = df[df[col].notna()]
    return df

def convert_columns_to_numeric(df):
    return df.apply(pd.to_numeric)

def convert_timestamp_to_datetime(df):
    return df

def remove_duplicates(df):
    return df.drop_duplicates()
```

**Convert to parquet file format**

```python
def convert_to_parquet(df, file_name):
    df.to_parquet(file_name)
```

**Program startup**

```python
def main():
    df_tickers_OHLCV_raw = get_data()
    df_tickers_OHLCV_clean = remove_duplicates(convert_timestamp_to_datetime(convert_columns_to_numeric(remove_NA(df_tickers_OHLCV_raw))))
    #df_tickers_OHLCV_clean.to_csv("test_data.csv")
    #convert_to_parquet(df_tickers_OHLCV_clean, "clean_prices.parquet")
main()
```
