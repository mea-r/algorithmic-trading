```python
import requests
import pandas as pd

def main():
    with open("api_keys/alpha_vantage.txt", "r") as f:
        api_key = f.readline().strip()

    ticker = "WMT"
    time_from = "20260309T0000" ##2026 March 9 Midnight
    time_to = "20260310T0000"
    limit = 50
    url = f"https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={ticker}&time_from={time_from}&time_to={time_to}&limit={limit}&apikey={api_key}"
    data = requests.get(url).json()

    news = data["feed"]
    headlines = [headline["title"] for headline in news]
    timestamps = [timestamp["time_published"] for timestamp in news]
    sources = [source["source"] for source in news]

    for headline in headlines:
        print(headline)

    for timestamp in timestamps:
        print(timestamp)

    for source in sources:
        print(source)
    #print(len(set(headlines)))
    #date = data["feed"][0]["time_published"]
    #print(pd.to_datetime(date, format = "mixed"))

main()
```
