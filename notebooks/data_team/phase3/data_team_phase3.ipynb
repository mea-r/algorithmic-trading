**Import relevant libraries**

```python
import requests
```

**Use Tiingo API to pull the most recent news for tickers**

```python
def get_data(tickers_file, key_file):
    with open(key_file, "r") as f:
        api_key = f.readline()
    with open(tickers_file, "r") as f:
        tickers = f.readline()
        tickers = tickers[1:-1]
        tickers = tickers.replace('"', "")
        tickers = tickers.split(",")
    tickers_joined = ",".join(tickers)
    tickers_joined = tickers_joined.replace(" ", "")
    #print(tickers_joined)
    headers = {'Content-Type': 'application/json', 'Authorization' : 'Token ' + api_key}
    print(headers)
    site = "https://api.tiingo.com/tiingo/news?tickers=aapl"
    print(site)
    requestResponse = requests.get(site, headers = headers)
    print(requestResponse.json())
```

**Main program**

```python
def main():
    get_data("tickers.txt", "Tiingo_api_key.txt")

main()
```
